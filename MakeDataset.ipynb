# Dataset Creation
This notebooks is the first step in the dataset creation. In this notebook I split the unstructured text into sentences..

The CNN/Daily Mail dataset is commonly used for abstractive summarization modeling

```python
# load the CNN/Daily Mail Dataset from HuggingFace Repo
dataset = load_dataset("cnn_dailymail", '3.0.0')
```

```python
# parse the articles into lists of sentences
new_dataset = dataset.map(lambda x: {"article_sent": [sent_tokenize(o) for o in x["article"]]}, batched=True)
```

```python
# get the number of sentences
new_dataset = new_dataset.map(lambda x: {"article_sent_len": len(o) for o in x["article_sent"]})
```

```python
# parse the sentences into words
# so now we have lists of lists of words
new_dataset = new_dataset.map(lambda x: {"article_sent_words":[y.split() for y in x["article_sent"]]})
```
